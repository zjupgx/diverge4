```python
from Bio import SeqIO
from Bio.Align import MultipleSeqAlignment
from Bio.Align.Applications import MuscleCommandline
from Bio.Seq import Seq
from io import StringIO
records = list(SeqIO.parse("ERBB_family_temp.fas", "fasta"))
sequences = [str(rec.seq).replace("-", "") for rec in records]
names = [rec.id for rec in records]

new_records = [SeqIO.SeqRecord(seq=Seq(sequences[i]), id=names[i],description="")  
                for i in range(len(sequences))]
                
with open("temp.fasta", "w") as output_handle:
    SeqIO.write(new_records, output_handle, "fasta")
```

```python
X_df = pd.DataFrame()
for tree_file in ["./EGFR.tree","./ERBB2.tree","./ERBB3.tree","./ERBB4.tree"]:
  rvs = Rvs("./ERBB_family_new.fas",tree_file)
  X = rvs.results().iloc[:,0]
  X_df[X.name] = X
```

```python
X_df.columns = ["X1","X2","X3","X4"]
X_df
```

令L是aligned stes的总数(不包含gap位置)

```python
X1=X_df.iloc[:,0]
X2=X_df.iloc[:,1]
X3=X_df.iloc[:,2]
X4=X_df.iloc[:,3]

L=X_df.shape[0]
A1 = sum(X1)/L
A2 = sum(X2)/L
A3 = sum(X3)/L
A4 = sum(X4)/L
B= sum(X1*X2*X3*X4) / L
C = sum((X1**2-X1)*(X2**2-X2)*(X3**2-X3)*(X4**2-X4))/L
R = C/(B*A1*A2*A3*A4)
```

在NCBI查找human和mouse的cds序列,计算dn/ds,令$\left(\frac{d N}{d s}\right)_i=\omega_i$
根据公式$\mu_c=-k \sum_{i=1}^4 \omega_i-k \ln R$计算得：$\mu_c=5.13$

```python
w1=0.068
w2=0.115
w3=0.086
w4=0.025
k=0.3223
Uc = -k*(w1+w2+w3+w4)-k*np.log(R)
print("μc=",Uc)
```
